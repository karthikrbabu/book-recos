```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
books_comm = pd.read_csv('../data/raw_data/popular_and_similar_books.csv')
books_comm.drop(books_comm.columns[0], axis=1, inplace = True)
```

```python
#books_comm['rank'==1]
books_comm.title.value_counts()
```

```python
b1 = books_comm[books_comm['book_id'] == 4410]
b2 = books_comm[books_comm['book_id'] == 1788]
union = pd.concat([b1,b2])
union
```

```python
title=[]
similar_title=[]
score=[]
title = b1['title'].to_list()
similar_title = b1['similar_title'].to_list()
score = b1['score'].to_list()
#setting plot size
plt.figure(3,figsize=(12,12)) 
G = nx.Graph()
#adding graph edges with start nodes and weighted edges between them 
for i in range(len(b1)):
    G.add_edge(title[i], similar_title[i], weight=score[i])
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size = 50,  with_labels = True, node_color = 'y')
nx.draw_networkx_edges(G, pos, alpha=.5)
nx.draw_networkx_labels(G,pos, font_size=9,font_family='sans-serif')
```
